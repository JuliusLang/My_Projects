## a) Write a programme that generates a discrete-time random walk Xt for t = 0, 1, 2, . . . with 200 steps. Take the following values: X0 = 0 and at each step t = 1, 2, . . ., Xt = Xt−1 + 1 with probability 21 , and Xt = Xt−1 − 1 with probability 12.
 

```python
def random_walk_a():
    """Generate and plot a discrete-time random walk X[t] for t=[0,200]"""

    X = list(range(0, 201)) # Created the list of values from 0 to 200 inclusively.

    for t in range(1, 201):
        prob = random.choice([1, -1]) # We choose randomly between 1 and -1 with the probability 0.5
        if prob == 1:
            X[t] = X[t-1] + 1
        else:
            X[t] = X[t-1] - 1
    
    plt.plot(X) # Plotting the graph.
    plt.show()

    #return X # Returning a value in case it is used not only in a plot, but other purposes.
random_walk_a()
```

## Programming Exercise  *(b)*

## Write a programme that takes in a parameter p ∈ [0, 1], and generates a similar random walk that goes “up” with probability p and down with probability 1 − p.

```python
def random_walk_b(p):
    """Generate and plot a discrete-time random walk that increases each step by 1
    with the probability p and decreases by 1 with the probability 1-p"""
    
    # First, we put constraints on the value of 'p'. If the user specifies the value > 1, the 'p' will still be 1.
    # If a user provides a negative number, the 'p' will be 0.
    if p < 0:
        p = 0
    elif p > 1:
        p = 1

    X = list(range(0,201)) # Created the list of values from 0 to 200 inclusively.

    for t in range(1, 201):
        prob = random.choices([1, -1], weights=[p, 1-p])[0] # We choose randomly between 1 and -1 with the probability 0.5
        if prob == 1:
            X[t] = X[t-1] + 1
        else:
            X[t] = X[t-1] - 1
    
    plt.plot(X) # Plotting the graph.
    plt.show()

    return X 


# random_walk_a() # random walks that change the value of X randomly with prob. 0.5 as plotted on a graph
random_walk_b(p=0.7) # random walks with the probability of growth 0.9, so the graph moves up most of the time
random_walk_b(p=0.3) # random walks with the probability of growth 0.1, so the graph will decrease most of the time

```

## Programming Exercise  *(c)*

## Write a programme that generates a Brownian motion Bt on [0, 1] Discretize the interval into 200 intervals of length ∆t := 1/200 , and define ti = i∆t for i = 0,1,...,200. Take Bt = 0 and at every iteration i = 1, 2, . . . , 200 generate a new, independent standard normal variable Zi and set
## B_ti = B_(ti−1) + √(∆t)Zi. 
## Present a graph with three different sample paths.

```python
def random_walk_c():
    """Generate a Brownian motion on [0,1] for 200 equal intervals of length delta_t = 1/200."""
    B_values = np.linspace(0,1,200) # array of 200 values of B (200 'delta t' intervals of length 1/200) evenly spaced between 0 and 1
    delta_t = B_values[1]

    Z = np.random.standard_normal(200)

    for i in range(1, 200):
        B_values[i] = B_values[i-1] + sqrt(delta_t) * Z[i-1]
    return B_values


# Generating 3 simultations of the Brownian motion on [0,1]
c_walk_1 = random_walk_c()
c_walk_2 = random_walk_c()
c_walk_3 = random_walk_c()

# Plotting 3 simulations on one graph.
plt.plot( c_walk_1, label='Simulation 1')
plt.plot( c_walk_2, label='Simulation 2')
plt.plot( c_walk_3, label='Simulation 3')

plt.title('3 simultations of the Brownian motion on [0,1]')
plt.xlabel('Number of Δt intervals')
plt.ylabel('Random step value')
plt.legend()
plt.show()



```

## Programming Exercise  *(d)*

## Now generalize this: write a programme that for any given input T > 0 and natural number N generates a Brownian motion on [0,T], discretized over N intervals of length NT . Present some graphs with different sample paths.

```python
def random_walk_d(T, N):
    """Generate a Brownian motion on [0,T] for N equal intervals of length delta_t = T/N."""
    B_values = np.linspace(0,T,N) # array of N values of B (T 'delta t' intervals of length T/N) evenly spaced between 0 and T
    delta_t = B_values[1]

    Z = np.random.standard_normal(N)

    for i in range(1, len(B_values)):
        B_values[i] = B_values[i-1] + sqrt(delta_t) * Z[i-1]
    return B_values


# Generating 3 simultations of the Brownian motion for T = 10000, N = 50000 so there will be 5000 intervals (Δt) between 0 and 10000
T = 10000
N = 50000
d_walk_1 = random_walk_d(T, N)
d_walk_2 = random_walk_d(T, N)
d_walk_3 = random_walk_d(T, N)

# Plotting 3 simulations on one graph.
plt.figure(figsize=(21, 12))
plt.plot(np.linspace(0,T,N),d_walk_1, label='Simulation 1')
plt.plot(np.linspace(0,T,N),d_walk_2, label='Simulation 2')
plt.plot(np.linspace(0,T,N),d_walk_3, label='Simulation 3')

plt.title('3 simultations of the Brownian motion on [{},{}]'.format(0,T))
plt.xlabel('Number of Δt intervals')
plt.ylabel('Random step value')
plt.legend()

plt.show()
```
